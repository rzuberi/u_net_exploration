```python
import tifffile
def import_images(images_path,num_imgs=20):
    images = [np.squeeze(tifffile.imread(images_path + str(i) + '.tiff')) for i in range(num_imgs)]
    return images

images = import_images(images_path)
imgs = [(image-np.min(image))/(np.max(image)-np.min(image)) for image in images]
```

```python
import cellpose
from cellpose import models

image = imgs[0]

cellpose_model = models.Cellpose(gpu=True, model_type='nuclei')
#make a prediction with this cellpose model
prediction = cellpose_model.eval(image, channels=[0,0])[0]
print(np.max(prediction))
```

```python
#binarise both predictions and calculate the dice loss between them

cellpose_binary_prediction = np.where(prediction>0,1,0)
print(cellpose_binary_prediction.shape)
model_binary_prediction = np.where(instance_masks[0]>0,1,0)
dice_loss = np.sum(cellpose_binary_prediction*model_binary_prediction)*2.0 / (np.sum(cellpose_binary_prediction) + np.sum(model_binary_prediction))
print(len(np.where(cellpose_binary_prediction==1)[0]))
print(len(np.where(model_binary_prediction==1)[0]))
print(dice_loss)
```
